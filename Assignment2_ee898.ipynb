# Creating XML to Dataframe

```python
class XMLParser:
    def __init__(self, xml_file):
        self.xml_file = xml_file
        self._root = ET.parse(self.xml_file).getroot()
        self._objects = self._root.findall("object")
        self.img_path = os.path.join(image_dir, self._root.find('filename').text)
        print(f"Image path: {self.img_path}")
        self.image_id = os.path.splitext(os.path.basename(self.img_path))[0]
        self.names = self._get_names()
        self.boxes = self._get_bndbox()

    def _get_names(self):
        names = []
        for obj in self._objects:
            name = obj.find("name")
            names.append(name.text)
        return names

    def _get_bndbox(self):
        boxes = []
        for obj in self._objects:
            coordinates = []
            bndbox = obj.find("bndbox")
            coordinates.append(int(bndbox.find("xmin").text))
            coordinates.append(int(bndbox.find("ymin").text))
            coordinates.append(int(bndbox.find("xmax").text))
            coordinates.append(int(bndbox.find("ymax").text))
            boxes.append(coordinates)
        return boxes

def xml_files_to_df(xml_files):
    names = []
    boxes = []
    image_id = []
    img_path = []
    for f in xml_files:
        xml = XMLParser(f)
        names.extend(xml.names)
        boxes.extend(xml.boxes)
        image_id.extend([xml.image_id] * len(xml.names))
        img_path.extend([xml.img_path] * len(xml.names))

    data = {
        "image_id": image_id,
        "class_name": names,
        "xmin": [box[0] for box in boxes],
        "ymin": [box[1] for box in boxes],
        "xmax": [box[2] for box in boxes],
        "ymax": [box[3] for box in boxes],
        "img_path": img_path
    }

    df = pd.DataFrame(data)
    df['label'] = df['class_name'].astype('category').cat.codes
    df = df.drop(columns=['class_name'])
    return df

def build_df(xml_files, img_dir):
    df = xml_files_to_df(xml_files)

    # Update the img_path to contain the full path to the images
    df['img_path'] = df['img_path'].apply(lambda x: os.path.join(image_dir, x))

    classes = sorted(df['label'].unique())
    return df, classes
```

# Funtion for displaying some image with bounding box

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from torchvision.transforms.functional import to_pil_image
import cv2

def show_images_with_boxes(df, image_dir, num_images=3):
    
    fig, axs = plt.subplots(1, num_images, figsize=(15, 5))

    for i in range(num_images):
        row = df.iloc[i]
        img_path = os.path.join(image_dir, row['img_path'])

        # Check if the image file exists
        if os.path.exists(img_path):
            img = cv2.imread(img_path)[..., ::-1]  

            labels = row['label']
            xmin = row['xmin']
            ymin = row['ymin']
            xmax = row['xmax']
            ymax = row['ymax']

            
            axs[i].imshow(img)
            axs[i].axis('off')

            
            rect = patches.Rectangle(
                (xmin, ymin), xmax - xmin, ymax - ymin,
                linewidth=1, edgecolor='r', facecolor='none')
            axs[i].add_patch(rect)
            axs[i].text(
                xmin, ymin, f"Class {labels}", color='r',
                backgroundcolor='w', fontsize=10)
        else:
            print(f"Image not found: {img_path}")

    plt.show()
```
